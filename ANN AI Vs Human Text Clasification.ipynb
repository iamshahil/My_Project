```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
```

```python
from sklearn.preprocessing import LabelEncoder
le1=LabelEncoder()
le2=LabelEncoder()
data['text']=le1.fit_transform(data['text'])
data['generated']=le2.fit_transform(data['generated'])
```

```python
x=data.drop(["generated"],axis=1)
x
```

```python
y=data["generated"]
y
```

```python
data["generated"].value_counts()
```

```python
xtrain,xtest,ytrain,ytest,=train_test_split(x,y,train_size=0.3,random_state=42)
```

```python
scaler = MinMaxScaler()
xtrain_scaled = scaler.fit_transform(xtrain)
xtest_scaled = scaler.transform(xtest)
xtrain_scaled
```

```python
xtrain_scaled.shape
```

```python
import tensorflow as tf
ann=tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(6,activation="relu"))
ann.add(tf.keras.layers.Dense(3,activation="relu"))
ann.add(tf.keras.layers.Dense(1,activation="sigmoid"))
#compiling Ann
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
#fit
ann.fit(xtrain_scaled,ytrain,batch_size=32,epochs=100)
```

```python
ypred = ann.predict(xtest_scaled)
ypred
```

```python
ypred_classes=ypred.argmax(axis=1)
ypred_classes
```

```python
prediction=ann.predict(scaler.transform([[83266]]))
if prediction>0.5:
    print("Generated by AI")
else:
    print("Generated by Human")
print(prediction)
```

```python
ann.summary()
```

```python
ann.evaluate(xtest_scaled,ytest)
```
