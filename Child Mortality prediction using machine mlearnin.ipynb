```python
from sklearn.preprocessing import LabelEncoder
obj= LabelEncoder()
OBJ = LabelEncoder()
```

```python
data_1['Entity']=obj.fit_transform(data_1['Entity'])
data_1['Code']=OBJ.fit_transform(data_1['Code'])
data_1
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = data_1.drop(['Under-fivemortalityrate'],axis=1)
y = data_1['Under-fivemortalityrate']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size =0.3,random_state = 42)
```

```python
xtest
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)*100
```
