```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv(r"Banglore_traffic_Dataset2.csv")
data
```

```python
data.dtypes
```

## EXPLORATORY DATA ANALYSIS

```python
data.isna().any().any()
```

```python
data.info()
```

```python
data.drop(["Date"],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.tail()
```

```python
data.describe()
```

```python
lb1=LabelEncoder()
lb2=LabelEncoder()
lb3=LabelEncoder()
lb4=LabelEncoder()
```

```python
# data['Date']=lb1.fit_transform(data['Date'])
data['AreaName']=lb1.fit_transform(data['AreaName'])
data['RoadIntersectionName']=lb2.fit_transform(data['RoadIntersectionName'])
data['WeatherConditions']=lb3.fit_transform(data['WeatherConditions'])
data['RoadworkandConstructionActivity']=lb4.fit_transform(data['RoadworkandConstructionActivity'])
data
```

```python
data['AreaName'] = pd.to_numeric(data['AreaName'], errors='coerce')
data['RoadIntersectionName'] = pd.to_numeric(data['RoadIntersectionName'], errors='coerce')
data['TrafficVolume'] = pd.to_numeric(data['TrafficVolume'], errors='coerce')
data['AverageSpeed'] = pd.to_numeric(data['AverageSpeed'], errors='coerce')
data['TravelTimeIndex'] = pd.to_numeric(data['TravelTimeIndex'], errors='coerce')
data['CongestionLevel'] = pd.to_numeric(data['CongestionLevel'], errors='coerce')
data['RoadCapacityUtilization'] = pd.to_numeric(data['RoadCapacityUtilization'], errors='coerce')
data['IncidentReports'] = pd.to_numeric(data['IncidentReports'], errors='coerce')
data['EnvironmentalImpact'] = pd.to_numeric(data['EnvironmentalImpact'], errors='coerce')
data['PublicTransportUsage'] = pd.to_numeric(data['PublicTransportUsage'], errors='coerce')
data['TrafficSignalCompliance'] = pd.to_numeric(data['TrafficSignalCompliance'], errors='coerce')
data['PedestrianandCyclistCount'] = pd.to_numeric(data['PedestrianandCyclistCount'], errors='coerce')
data['WeatherConditions'] = pd.to_numeric(data['WeatherConditions'], errors='coerce')
data['RoadworkandConstructionActivity'] = pd.to_numeric(data['TrafficSignalCompliance'], errors='coerce')
														
data.dtypes
```

```python
x=data.drop(['TrafficVolume'],axis=1)
y=data['TrafficVolume']
```

```python
x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
xtrain
```

```python
ytrain
```

## LinearRegression

```python
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
```

```python
model1.fit(xtrain,ytrain)
```

```python
LRtrain=model1.score(xtrain,ytrain)*100
LRtrain
```

```python
LRtest=model1.score(xtest,ytest)*100
LRtest
```

## DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
model2=DecisionTreeRegressor()
```

```python
model2.fit(xtrain,ytrain)
```

```python
DTtrain=model2.score(xtrain,ytrain)*100
DTtrain
```

```python
DTtest=model2.score(xtest,ytest)*100
DTtest
```

## RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model3=RandomForestRegressor(n_estimators=100)
```

```python
model3.fit(xtrain,ytrain)
```

```python
RFtrain=model3.score(xtrain,ytrain)*100
RFtrain
```

```python
RFtest=model3.score(xtest,ytest)*100
RFtest
```

## AdaBoostRegressor

```python
from sklearn.ensemble import AdaBoostRegressor
model4=AdaBoostRegressor()
```

```python
model4.fit(xtrain,ytrain)
```

```python
ABtrain=model4.score(xtrain,ytrain)*100
ABtrain
```

```python
ABtest=model4.score(xtest,ytest)*100
ABtest
```

## GradientBoostingRegressor

```python
from sklearn.ensemble import GradientBoostingRegressor
model5=GradientBoostingRegressor(n_estimators=100,random_state=1)
```

```python
model5.fit(xtrain,ytrain)
```

```python
GBtrain=model5.score(xtrain,ytrain)*100
GBtrain
```

```python
GBtest=model5.score(xtest,ytest)*100
GBtest
```

## KNeighborsRegressor

```python
from sklearn.neighbors import KNeighborsRegressor
model6= KNeighborsRegressor(n_neighbors=5)
model6.fit(xtrain,ytrain)
```

```python
KNtrain=model6.score(xtrain,ytrain)*100
KNtrain
```

```python
KNtest=model6.score(xtest,ytest)*100
KNtest
```
